### Advanced Task 1 A Fourth Client

Because the client does not know who the customers are, this is an unsupervised learning model.  
The ML task type that would best suite this question is clustering.  
The three types of clustering include:  
Centroid-based clustering - the analyst picks a number of groups, determines the center, and datapoints are assigned to the nearest value.  
Hierarchical clustering - Each data point is treated as a group and merged until there is one group. This allows you to chop the tree at various points to examine different levels of specificity.  
Density-based clutering - identify regionally compact groups  
I recommend the density based because of the lack of direction from the customer!
